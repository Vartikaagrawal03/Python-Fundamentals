```python
##create a set
my_set={1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
my_empty_set=set([1,2,3,4,5,6])
print(my_empty_set)
```

```python
my_empty_set=set([1,2,3,5,6,4,5,6])
print(my_empty_set)
```

```python
## basic set operations
#adding element
my_set.add(7)
print(my_set)
```

```python
##remove element
my_set.remove(3)
print(my_set)
```

```python
## pop method
removed_element=my_set.pop()
print(removed_element)
print(my_set)
```

```python
## clear all the elements
my_set.clear()
print(my_set)
```

```python
## set membership test
my_set={1,2,3,4,5}
print(3 in my_set)
print(10 in my_set)
```

```python
## mathematical operation 
set1={1,2,3,4,5}
set2={3,4,5,6,7}

#union
union_set=set1.union(set2)
print(union_set)

#intersection
intersection_set=set1.intersection(set1)
print(intersection_set)
```

```python
set1={1,2,3,4,5}
set2={3,4,5,6,7}

## difference
print(set1.difference(set2))
```

```python
## set methods
set1={1,2,3,4,5}
set2={3,4,5}

## is subset
print(set1.issubset(set2))

## is superset
print(set1.issuperset(set2))
```

```python
## counting unique words in text

text="My name is vartika agrawal"
words=text.split()

##convert list of words to set to get unique words
unique_words=set(words)
print(unique_words)
print(len(unique_words))
```
