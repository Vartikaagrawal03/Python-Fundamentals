```python
import pandas as pd
```

```python
 ## series
 ## pandas series is a one dimensional array like object that can hold any data type.

 data=[1,2,3,4,5]
 series=pd.Series(data)
 print("series\n",series)
```

```python
## create a series from dictionary element
data={'a':1,'b':2,'c':3}
series_dict=pd.Series(data)
print(series_dict)
```

```python
data=[10,20,30]
index=['a','b','c']
pd.Series(data,index=index)
```

```python
##Dataframe
## create a dataframe from dictionary

data={
    'Name':['Krish','Vartika','Sambhava'],
    'Age':[25,20,20],
    'City':['Bangalore','Kota','Jaipur']
}
df=pd.DataFrame(data)
print(df)
print(type(df))
```
